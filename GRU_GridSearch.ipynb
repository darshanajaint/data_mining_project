```python
learning_rates = [2*(10**-5)]
input_sizes = [50, 100, 200]
hidden_sizes = [50, 100, 300]
dropout = [0, 0.2, 0.4]
bi = [True, False]
```

```python
epochs = 25
batch_size = 128
train_csv = 'train_data.csv'
test_csv = 'test_data.csv'
model_base = './models_gridsearch/'
metrics_base = './metrics_gridsearch/'
```

```python
!mkdir models_gridsearch
!mkdir metrics_gridsearch
```

```python
for lr in learning_rates:
  for input in input_sizes:
    for hidden in hidden_sizes:
      for drop in dropout:
        for direction in bi:
          models_path = '{:.4f}_{:d}_{:d}_{:.1f}_{:s}'.format(lr, input, hidden, drop, str(direction))
          metrics_path = models_path + '.pt'
          !python pytorch_training.py --batch_size $batch_size --train_csv $train_csv  \
            --lr $lr --epochs $epochs --input_size $input --hidden_size $hidden  \
            --dropout $drop --bidirectional $direction --model_save_path $models_path  \
            --metrics_save_path $metrics_path --test_after_train False

      current_iter = '{:.4f}_{:d}_{:d}.zip'.format(lr, input, hidden)

      zip_models = 'models_' + current_iter
      base_models = './models_gridsearch/'

      zip_metrics = 'metrics_' + current_iter
      base_metrics = './metrics_gridsearch/'

      !zip -r $zip_models $base_models
      !zip -r $zip_metrics $base_metrics 

      drive_base_path = '/content/drive/My\ Drive/Spring\ 2021/Data\ Mining\ Sp2021/results/'

      drive_models_path = drive_base_path + 'models_gridsearch/'
      !cp $zip_models $drive_models_path

      drive_metrics_path = drive_base_path + 'metrics_gridsearch/' 
      !cp $zip_metrics $drive_metrics_path
```
